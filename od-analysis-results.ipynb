```python
import pandas as pd
import numpy as np
# visualization
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid', context='notebook')
%matplotlib notebook
# %matplotlib inline
```

```python
aucroc_unsup = pd.read_csv('./aucroc_unsup.csv')
aucroc_unsup.head()
```

```python
aucroc_semi = pd.read_csv('./aucroc_semi.csv')
aucroc_semi.head()
```

```python
aucpr_unsup = pd.read_csv('./aucpr_unsup.csv')
aucpr_unsup.head()
```

```python
aucpr_semi = pd.read_csv('./aucpr_semi.csv')
aucpr_semi.head()
```

```python
ds_dict = {'dataset':['cardio', 'musk', 'optdigits', 'speech', 'vowels']}
df_ds = pd.DataFrame(ds_dict)
df_ds
```

```python
aucroc_unsup_modif = aucroc_unsup[['DeepSVDD','DAGMM', 'COPOD', 'ECOD']]
aucroc_semi_modif = aucroc_semi.iloc[:,1:]
aucroc_final = pd.concat([df_ds, aucroc_unsup_modif, aucroc_semi_modif], axis=1)
aucroc_final
```

```python
aucpr_unsup_modif = aucpr_unsup[['DeepSVDD','DAGMM', 'COPOD', 'ECOD']]
aucpr_semi_modif = aucpr_semi.iloc[:,1:]
aucpr_final = pd.concat([df_ds, aucpr_unsup_modif, aucpr_semi_modif], axis=1)
aucpr_final
```

```python
aucroc_final_index = aucroc_final.set_index('dataset')
aucroc_final_index
```

```python
aucpr_final_index = aucpr_final.set_index('dataset')
aucpr_final_index
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(aucroc_final_index, 
        xticklabels=aucroc_final_index.columns,
        yticklabels=aucroc_final_index.index, annot=True, ax=ax)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(aucpr_final_index, 
        xticklabels=aucpr_final_index.columns,
        yticklabels=aucpr_final_index.index, annot=True, ax=ax)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
