```python
import pandas as pd
import numpy as np
# visualization
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid', context='notebook')
%matplotlib notebook
# %matplotlib inline
```

```python
df = pd.read_csv('./datasets/info_dataset.csv', index_col=None)
df.head()
```

```python
cat = pd.DataFrame(df.Category.value_counts())
```

```python
cat = cat.reset_index()
cat = cat.rename(columns={'index':'Category', 'Category':'Count'})
```

```python
cat
```

# Category

```python
fig, ax = plt.subplots(figsize=(10,6))

sns.barplot(data=cat, y='Category', x='Count')
plt.xlabel('Count')
plt.ylabel('Category')
plt.title("Dataset by Category")
# plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))

plt.pie(cat.Count, labels=cat.Category , autopct='%.0f%%')
plt.title("% Dataset by Category")
# plt.legend()
plt.show()
```

# Features
